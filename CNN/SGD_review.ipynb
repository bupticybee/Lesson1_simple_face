# SGD

```python
y = (x - 1) ** 2
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
xs = []
ys = []
for x in np.arange(-1,4,0.05):
    xs.append(x)
    ys.append((x - 1) ** 2)
```

```python
plt.scatter(xs,ys)
```

```python
lr = 0.01
x = 4
log = []
for i in range(300):
    grad = 2 * x - 2
    x = x - grad * lr
    log.append(x)
```

```python
import pandas as pd
```

```python
pd.DataFrame(log).plot()
```

```python
lr = 0.01
x = np.random.randn(10)
log = []
for i in range(300):
    grad = 2 * x - 2
    x = x - grad * lr
    log.append(x)
```
